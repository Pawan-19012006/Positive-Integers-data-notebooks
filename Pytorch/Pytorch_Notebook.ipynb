#Advanced Slicing

```python
indices = torch.tensor([0,2]) #when dim = 0, we get to select the rows and the torch.tensor([0,2]) is necessary, we can either put that into the index or we need to store it in a seperate variable and then use it
selected_rows = torch.index_select(twod,dim=0,index=indices)
selected_rows
```

```python
indices = torch.tensor([0,2]) #when dim=1, we get to select the cols
selected_rows = torch.index_select(twod,dim=1,index=indices)
selected_rows
```

```python
twod[:,::2] # Here we use Step, ie, we select all the rows, then start from the first col, the skip one col, print the next col, hence it has printed zeroth and last column
```

```python
torch.flip(twod,dims=(0,)) #This will flip all the rows
```

```python
print(twod) #Flips all the columns
twod=torch.flip(twod,dims=[1])
twod
```

#Memory Allocation

```python
float16_tensor = torch.ones(1000, dtype=torch.float16)
float32_tensor = torch.ones(1000, dtype=torch.float32)
float64_tensor = torch.ones(1000, dtype=torch.float64)

print(float16_tensor.element_size())
print(float32_tensor.element_size())
print(float64_tensor.element_size())
print(float16_tensor.nelement())
print("Memory taken by float16 :", float16_tensor.element_size()*float16_tensor.nelement())
print("Memory taken by float32 :", float32_tensor.element_size()*float32_tensor.nelement())
print("Memory taken by float64 :", float64_tensor.element_size()*float64_tensor.nelement())
```

#Concatenation of Tensors (Joining Tensors)

First we use .cat which allows the joining of tensors without altering the dimensions of the tensors

```python
tensor1 = torch.tensor([[1,2],[3,4]])
tensor2 = torch.tensor([[5,6],[7,8]])
joined_tensor_row = torch.cat((tensor1,tensor2),dim=0)
joined_tensor_col = torch.cat((tensor1,tensor2),dim=1)
print(joined_tensor_row)
print(joined_tensor_col)
```

Next we use .stack which will modify the dimensions of the tensors

```python
joined_tensor_row_stack = torch.stack((tensor1,tensor2),dim=0)
joined_tensor_col_stack = torch.stack((tensor1,tensor2),dim=1)
print(joined_tensor_row_stack)
print(joined_tensor_col_stack)
```

#Splitting of Tensors

First we use .chunk to split the tensors into equal sized chunks, it will split it into n number of chunks, but all the chunks will have same number of elements

```python
tensor = torch.arange(12)
chunked_tensor = torch.chunk(tensor,6,dim=0)
chunked_tensor
```

Here even though we give 5, the split is only happened into 4 chunks, this is becoz, when split in 5, all chunks will not have same number of elements, hence it will look for the previous vals where the split is equal and take that value as default, this wont happen in .split

```python
chunked_tensor_2 = torch.chunk(tensor,5,dim=0)
chunked_tensor_2
```

Whereas if we use split, then the split will happen based on the number of elements, ex, if 5, then each part will have 5 elements in it

```python
split_tensor = torch.split(tensor,5,dim=0)
split_tensor
```

#Reshaping a Tensor

```python
normal_tensor = torch.arange(24)
tensor_reshape = torch.arange(24).reshape(12,2)
print(normal_tensor)
tensor_reshape
```

#Transposing

```python
transpose_tensor = tensor_reshape.transpose(0,1) #When we transpose 0 and 1, we actually have the dimensions as (12,2), here it will get swapped into (2,12)
print(tensor_reshape)
print(transpose_tensor)
```
